```python
import random
import matplotlib.pyplot as plt
import numpy as np
import time
```

$\textbf{Problem 1}$

$\textbf{Part a)}$

```python
# Part a)

def rejection():
    while True:
        x = np.random.uniform(-1, 1, 2600)
        y = np.random.uniform(-1, 1, 2600)
        
        mask = x**2 + y**2 <= 1
        
        valid_x = x[mask]
        valid_y = y[mask]
        
        if len(valid_x) >= 2000:
            return valid_x[:2000], valid_y[:2000]

start = time.time()
x_1, y_1 = rejection()
end = time.time()
print("Elapsed time: ", end-start)

plt.figure(figsize=(6,6))
plt.scatter(x_1, y_1, s=2)
plt.gca().set_aspect('equal')
plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5])
plt.title("Points uniformly distributed on a circle with radius 1 with the rejection method")
plt.show()
```

$\textbf{Part b)}$

```python
# Part b)

def polar_coordinates():
    theta = np.random.uniform(0, 1, 2000)
    r = np.random.uniform(0, 1, 2000)
    ret_x = r * np.cos(2 * np.pi * theta)
    ret_y = r * np.sin(2 * np.pi * theta)
    
    return ret_x, ret_y

start = time.time()
x_2, y_2 = polar_coordinates()
end = time.time()
print("Elapsed time: ", end-start)

plt.figure(figsize=(6,6))
plt.scatter(x_2, y_2, s=2)
plt.gca().set_aspect('equal')
plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5])
plt.title("Points uniformly distributed on a circle with radius 1 with polar coordinates")
plt.show()
```

We notice that the center of the circle seems to be denser than the points further away from the center of the circle.
This is due to the fact that sampling the radius from a uniform distribution makes the center denser than it should be.
We need more points near the border since for a small change in the angle theta the area "sweeped" by this change is
greater the further away you are from the center.

$\textbf{Part c)}$

$\textbf{i)}$

$$
( p(r) = C 2 \pi r \wedge F(r=R) = 1 ) \Rightarrow \int_0^1 p(r) dr = 1 
$$

$$
\Rightarrow \left. C \pi r^2 \right|_{0}^{1} = 1 \Rightarrow C = \frac{1}{\pi}
$$

$$
\text{So we get: } r = F^{-1}(y) = \sqrt{y}
$$


$\textbf{ii)}$

```python
# Part c)

# if p(r) = C*2*pi*r and it's integral from 0 to 1 must be one, we get that C*pi*r^2 evaluated at r = 1 minus it evaluated
# at r = 0 must be equal to one since it is a probability distribution. From this we get that C * pi * 1^2 = 1 => C = 1 / pi
# if the cumulative distribution for r is then r^2, its inverse is sqrt(r) and this is then the transformation we will be applying
# to the uniformly distributed r

C = 1 /np.pi
def inverse_distribution_method():
    theta = np.random.uniform(0, 1, 2000)
    r = np.random.uniform(0, 1, 2000)
    r = np.sqrt(r)
    ret_x = r * np.cos(2 * np.pi * theta)
    ret_y = r * np.sin(2 * np.pi * theta)
    
    return ret_x, ret_y

start = time.time()
x_3, y_3 = inverse_distribution_method()
end = time.time()
print("Elapsed time: ", end-start)

plt.figure(figsize=(6,6))
plt.scatter(x_3, y_3, s=2)
plt.gca().set_aspect('equal')
plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5])
plt.title("Points uniformly distributed on a circle with radius 1 with the inverse distribution method")
plt.show()
```

$\textbf{Part d)}$

We can clearly see that the inverse distribution method proved to result in a much more uniform distribution on the unit circle.
The difference comes from choosing a proper distribution to sample the radius from due to the fact that uniformly distributed
radii results in a much more dense center, which is obvious from the nature of polar coordinates and how the area sweeped
by a small change in the angle theta is much larger the further away you are from the center. This distortion should be familiar
from calculus and it is the reason why we multiply by r when we decide to switch from cartesian coordinates to 
polar coordinates in integration.

$\textbf{Problem 2}$

$\textbf{Part a)}$

$$
f(x) = C x^{-3/2}
$$

$$
\int_1^\infty f(t) dt = 1 \Leftrightarrow \int_1^\infty C t^{-3/2} dt = 1
$$

$$
-2 C t^{-1/2} |_1^\infty = 1 \Leftrightarrow 0 - (-2 C \cdot 1^{-1/2}) = 1 \Leftrightarrow C = 1/2
$$

Hence we got that C = 1/2

$$
F(x) = \int_1^x f(t) dt = \int_1^x 1/2 t^{-3/2} = -t^{-1/2} |_1^x = (-x^{-1/2}) - (-1) = 1 - x^{-1/2}
$$

$$ 
y = F(x) \Leftrightarrow y = 1 - x^{-1/2} \Rightarrow x^{-1/2} = 1 - y \Rightarrow x = \frac{1}{(1-y)^2} \Leftrightarrow x = F^{-1}(y) 
$$

So $x = F^{-1}(y) = \frac{1}{(1-y)^2}$ and therefore $F^{-1}(y/A) = \frac{1}{(1-y/A)^2}$

```python
def power_law(A=200000, B=190000):
    y = np.random.uniform(0, B, 1000)
    x = 1 / ((1 - y / A)**2)
    return x

data = power_law()
numberOfBins = 30
```

$\textbf{Part b)}$

```python
# part b) linear binning in linear coordinates

x = np.linspace(min(data), max(data), 10000)
y = 0.5*(x**(-3/2))
plt.hist(data, bins=np.linspace(0,350, numberOfBins), density='true', label='Data')
plt.title('Data in linear coordinates using linear binning')
plt.plot(x, y, 'r-', linewidth = 1, label='0.5*x^(-3/2)')
plt.legend()
plt.show()
```

```python
# part b) linear binning in logarithmic coordinates

plt.hist(data, bins=np.linspace(1,350, numberOfBins), density='true', label='Data')
plt.title('Data in logarithmic coordinates using linear binning')
plt.xscale('log')
plt.yscale('log')
plt.plot(x, y, 'r-', linewidth = 1, label='0.5*x^(-3/2)')
plt.legend()
plt.show()
```

$\textbf{Part c)}$

```python
# Part c) linearly binned data values at the midpoints in logarithmic coordinates

count, edges = np.histogram(data, bins=numberOfBins, density='true')

midpoints = (edges[:-1] + edges[1:]) / 2

plt.plot(midpoints, count, 'o-', label='Data')
plt.title('Data in logarithmic coordinates using midpoints of linear bins')
plt.xscale('log')
plt.yscale('log')
plt.plot(x, y, 'r-', linewidth = 1, label='0.5*x^(-3/2)')
plt.legend()
plt.show()
```

$\textbf{Part d)}$

```python
# Part d) Logarithmic binning in logarithmic coordinates

bins = np.logspace(np.log10(1), np.log10(400), numberOfBins)

midpoints = 0.5 * (bins[:-1] + bins[1:])

count, edges = np.histogram(data, bins=bins, density='true')

plt.plot(midpoints, count, 'o-', label='Data')
plt.title('Data in logarithmic coordinates using the midpoints of logarithmic bins')
plt.xscale('log')
plt.yscale('log')
plt.plot(x, y, 'r-', linewidth = 1, label='0.5*x^(-3/2)')
plt.legend()
plt.show()
```

$\textbf{Part e)}$

Doing logarithmic binning instead of linear binning in logarithmic coordinates in part d) vs part c) proved to be much more beneficial.
The logarithmic binning resulted in a much better plot with more even distances between the bin midpoints in logarithmic coordinates,
this makes identifying the underlying pdf easier in a real world application. A disadvantage for linear binning is the result of bins that are too narrow further away from the origin, this results in empty bins when data is scarce as we can see from the plot made in part c). The resulting plot made with midpoints of linear bins looks quite silly when plotted next to the actual pdf. The midpoints distort the actual sampled data.

$\textbf{Part f)}$

Simulating the pdf via the rejection method would prove really cumbersome due to the exponential nature of the pdf. Getting reliable results
far away from the origin would take very large number of trials and is in no way computationally efficient. This is due to the fact that
the pdf is exponential so it decays exponentially and since you need to sample from the whole real line (or more realistically a very
large interval e.g. $[1, 10 000]$, you would need a huge number of samples to simulate the pdf properly with the rejection method.
