```python
import random
import matplotlib.pyplot as plt
import numpy as np
import time
import math
```

$\textbf{Problem 1}$

```python
# Constants and initial values

a = 1
b = 1
c = 3
d = 1/100

mu_init = a / c
tau_init = c
```

```python
# The Gibbs sampling algorithm

def gibbs(mu_init, tau_init, a, b, c, d, N=100_000, n=20, s2=20, x_bar=30):
    mu = mu_init
    tau = tau_init
    mus = []
    taus = []
    alpha = a + n/2 # Stays constant through the run so no need to compute this repeatedly
    for i in range(N):
        #print("mu: ", mu)
        # Compute the parameters for the Gamma distribution of tau based on previous mu
        beta = b + 1/2 * ((n - 1) * s2 + n * (x_bar - mu)**2)
        tau = np.random.gamma(alpha, 1 / beta) # Gamma(k, theta), k = alpha, theta = 1/beta
        
        # Now compute the required parameters with the just computed tau for mu
        mean = (c * d + n * tau * x_bar) / (n * tau + d)
        stdev = math.sqrt(1 / (n * tau + d))
        
        # Update mu
        mu = np.random.normal(mean, stdev) # N(mean, stdev)
        
        mus.append(mu)
        taus.append(tau)
    return mus, taus
```

```python
mu, tau = gibbs(mu_init, tau_init, a, b, c, d)
```

```python
# i) Plot mu vs iteration steps
plt.figure()
plt.plot(mu)
plt.title('i) Mu vs Iteration Steps')
plt.xlabel('Iteration')
plt.ylabel('Mu')
plt.show()

# ii) Plot marginal distribution of mu
plt.figure()
plt.hist(mu, bins=50, density=True)
plt.title('ii) Marginal Distribution of Mu')
plt.xlabel('Mu')
plt.ylabel('Density')
plt.show()

# iii) Plot tau vs iteration steps
plt.figure()
plt.plot(tau)
plt.title('iii) Tau vs Iteration Steps')
plt.xlabel('Iteration')
plt.ylabel('Tau')
plt.show()

# iv) Plot marginal distribution of tau
plt.figure()
plt.hist(tau, bins=50, density=True)
plt.title('iv) Marginal Distribution of Tau')
plt.xlabel('Tau')
plt.ylabel('Density')
plt.show()

# v) Plot joint distribution of mu and tau
plt.figure()
plt.scatter(mu, tau, alpha=0.1)
plt.title('v) Joint Posterior Distribution of Mu and Tau')
plt.xlabel('Mu')
plt.ylabel('Tau')
plt.show()
```

```python
mean_mu = np.mean(mu) # vi)
mean_tau = np.mean(tau) # vii)
print(f"Mean of mu: {mean_mu}")
print(f"Mean of tau: {mean_tau}")
```
