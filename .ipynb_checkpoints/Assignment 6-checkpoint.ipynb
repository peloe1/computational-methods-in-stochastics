```python
import random
import matplotlib.pyplot as plt
import numpy as np
import time
import math
```

```python
alpha = 3
beta = 1

def gamma_density(x, alpha=alpha, beta=beta):
    return (x**(alpha-1)) * np.exp(-x * beta) * (beta ** alpha) / math.gamma(alpha)
    # Assuming integer alpha = 3,  beta = 1, Gamma(3) = (3-1)! = 2! = 2
    #return (x**2) * np.exp(-x) / 2

def exponential_density(x, lambda_i):
    return np.exp(- x * lambda_i) * lambda_i

def metropolis_hastings(lambda_i, niter):
    acceptances = 0
    thetas = np.zeros(niter + 1)
    theta = np.random.exponential(1 / lambda_i)
    thetas[0] = theta
    
    for i in range(1, niter + 1):
        #proposal = [np.random.exponential(lambda_i)]

        #nominator = product_of_gammas(proposal) * product_of_exponentials(proposal, lambda_i)
        #denominator = product_of_gammas(thetas) * product_of_exponentials(thetas, lambda_i)
        
        proposal = np.random.exponential(1 / lambda_i)
        
        nominator = gamma_density(proposal) * exponential_density(theta, lambda_i)
        denominator = gamma_density(theta) * exponential_density(proposal, lambda_i)
        
        acceptance_prob = min(1, nominator / denominator)

        if np.random.uniform(0, 1) <= acceptance_prob:
            theta = proposal
            acceptances += 1
            
        thetas[i] = theta
    
    thetas = np.delete(thetas, 0)
    return thetas, acceptances
```

```python
stepsize = 0.05
lambdas = np.arange(0.1, 1.0 + stepsize, stepsize)
niter = 10**5

acceptance_ratios = np.zeros(len(lambdas))
k = 0
for lambda_i in lambdas:
    _, acceptances = metropolis_hastings(lambda_i, niter)
    acceptance_ratios[k] = acceptances / niter
    k += 1
```

```python
lambda_opt = lambdas[np.argmax(acceptance_ratios)]
print("Optimal lambda: ", lambda_opt)
```

```python
plt.plot(lambdas, acceptance_ratios)

ticks = np.arange(0.1, 1.1, 0.1)
plt.xticks(ticks)

plt.xlabel('Lambda')
plt.ylabel('Acceptance Ratio')
plt.title('Acceptance Ratios vs. Lambda')
plt.show()
```

Looks like the optimal lambda is somewhere between 0.3 and 0.4, so let us do another search just on that interval with a smaller step size.

```python
stepsize = 0.01
lambdas = np.arange(0.3, 0.4 + stepsize, stepsize)
niter = 10**5

acceptance_ratios = np.zeros(len(lambdas))
k = 0
for lambda_i in lambdas:
    _, acceptances = metropolis_hastings(lambda_i, niter)
    acceptance_ratios[k] = acceptances / niter
    k += 1
```

```python
plt.plot(lambdas, acceptance_ratios)

ticks = np.arange(0.3, 0.4 + 0.01, 0.01)
plt.xticks(ticks)

plt.xlabel('Lambda')
plt.ylabel('Acceptance Ratio')
plt.title('Acceptance Ratios vs. Lambda')
plt.show()
```

Since this interval $[0.3, 0.4]$ has a width of double the tolerance 0.05 given in the assignment we can confidently pick any point on this interval to be our choice of the optimal lambda and it should be within tolerance of the true value. Based on the figure $\lambda_{opt} = 0.35$

```python
lambda_opt = lambdas[np.argmax(acceptance_ratios)]
print("Optimal lambda: ", lambda_opt)
```

And taking the maximum of the acceptance ratios does indeed correspond to a lambda of 0.35

Part b)

```python
simulated_distribution, _ = metropolis_hastings(lambda_opt, 10**6)
```

```python
plt.hist(simulated_distribution, bins=500, density='true', label="Simulated Gamma(3,1) distribution with M-H")

x_axis = np.arange(0,20,0.1)
true_gamma_distribution = [gamma_density(x) for x in x_axis]
plt.plot(x_axis, true_gamma_distribution, 'r', linewidth=0.75, label="True Gamma(3,1) distribution")
plt.legend()
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Comparison of Simulated and True Gamma(3,1) Distributions")
plt.show()
```

```python
# Mean and variance of the simulated distribution
mean = np.mean(simulated_distribution)
variance = np.var(simulated_distribution)

print("Mean :", mean)
print("Variance :", variance)
```
