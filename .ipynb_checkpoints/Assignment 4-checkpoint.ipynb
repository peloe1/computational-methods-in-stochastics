```python
import random
import matplotlib.pyplot as plt
import numpy as np
import time
import math
```

$\textbf{Problem 1}$

```python
lambdas = [0.2, 0.9, 0.5, 0.6, 1.0, 0.6, 0.5]
hours = [(0,6),(6,8),(8,12),(12,15),(15,17),(17,19),(19,24)]

U_lambda = max(lambdas)

def lambdaAtTime(t):
    for k in range(0, len(hours)):
        if hours[k][0] <= t and hours[k][1] > t:
            return lambdas[k]
    print("Error, something happened")
    return -1

def minutesToHours(minutes):
    hours = minutes / 60
    return hours
```

```python
def poissonGenerator(lmbda): # inverse transform approach adapted from Wikipedia
    x = 0
    p = np.exp(-lmbda)
    s = p
    u = random.random() # U(0,1) sample
    while u > s:
        x += 1
        p *= lmbda / x
        s += p
    return x
        

def revisedPoisson(T): # T is minutes
    x = 0 # float, minutes
    times = []
    #m = poissonGenerator(U_lambda * T) # this approach proved to be quite slow, but worked
    m = np.random.poisson(U_lambda * T)
    for i in range(1, m+1):
        u = random.random()
        x = x + (T - x) * (1 - u**(1/(m - i + 1)))
        y = random.uniform(0, U_lambda)
        if y <= lambdaAtTime(minutesToHours(x)):
            times.append(x)
    return times

def thinningPoisson(T): # T is minutes
    times = []
    t = 0
    i = 1
    while True:
        time_sample = -U_lambda * np.log(random.random())
        nextT = t + time_sample
        if nextT > T:
            break
        u = random.random()
        if u <= lambdaAtTime(minutesToHours(nextT)) / U_lambda:
            times.append(nextT)
        t = nextT
        i += 1
    return times
```

```python
def simulation(n, T): # T is minutes
    results_revised = []
    results_thinning = []
    for i in range(0,n):
        results_revised.append(revisedPoisson(T))
        results_thinning.append(thinningPoisson(T))
    return results_revised, results_thinning

def averagesPerMinute(n, results, T): # T is minutes
    averages = [0 for _ in range(0, T)] # initialize to zero for each minute of the day
    for i in range(0, n):
        for k in range(0, len(results[i])):
            averages[math.floor(results[i][k])] += 1
    
    return [x / n for x in averages]

def cumulativePerMinute(averages):
    cumulative = [averages[0]]
    for i in range(1, len(averages)):
        cumulative.append(cumulative[-1] + averages[i])
    return cumulative

def numberOfCarsPerDay(simulation):
    return len(simulation)

def timeUntilFirst(simulation):
    i = 0
    while simulation[i] == 0:
        i += 1
    return simulation[i]
```

```python
# The data

n = 1000
T = 24*60 # 24 hours in minutes
minutes = [i for i in range(0, T)]
revised, thinning = simulation(n, T)

averages_revised = averagesPerMinute(n, revised, T)
averages_thinning = averagesPerMinute(n, thinning, T)

cumulative_revised = cumulativePerMinute(averages_revised)
cumulative_thinning = cumulativePerMinute(averages_thinning)

averageNumberOfCars_revised = np.mean([numberOfCarsPerDay(x) for x in revised]) # Expectation is 750

averageNumberOfCars_thinning = np.mean([numberOfCarsPerDay(x) for x in thinning]) # Expectation is 750

averageTimeFirst_revised = np.mean([timeUntilFirst(x) for x in revised])
averageTimeFirst_thinning = np.mean([timeUntilFirst(x) for x in thinning])
```

$\textbf{Part a) The revised method}$

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12, 5))
ax1.scatter(minutes, averages_revised, s=1)
ax1.set_title("Revised method: Average number of cars per minute.")
ax1.set_xlabel("Time (minutes)")
ax2.scatter(minutes, cumulative_revised, s=1)
ax2.set_title("Revised method: Cumulative average number of cars.")
ax2.set_xlabel("Time (minutes)")

print("Average number of cars arrived during the day in the revised approach simulation: ", averageNumberOfCars_revised)
print("Average time until the first car arrived in the revised approach simulation: ", averageTimeFirst_revised, " minutes.")
```

$\textbf{Part b) The thinning approach}$

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12, 5))
ax1.scatter(minutes, averages_thinning, s=1)
ax1.set_title("Thinning approach: Average number of cars per minute.")
ax1.set_xlabel("Time (minutes)")
ax2.scatter(minutes, cumulative_thinning, s=1)
ax2.set_title("Thinning approach: Cumulative average number of cars.")
ax2.set_xlabel("Time (minutes)")
print("Average number of cars arrived during the day in the thinning approach simulation: ", averageNumberOfCars_thinning)
print("Average time until the first car arrived in the thinning approach simulation: ", averageTimeFirst_thinning, " minutes.")
```

$\textbf{Problem 2}$

$\textbf{Part a)}$

```python
N0 = 10_000
lambda_0 = 0.3
num_simulations = 1_000
delta_t = 0.01

def homogenous(N0, lambda_0, delta_t):
    count = 0
    steps = 0
    while count < N0 / 2:
        m = np.random.poisson(N0 * lambda_0 * delta_t) # Sample how many decays happened at this time step
        count += m # Increase the number of decays
        steps += 1 # Move to next time step
    return steps * delta_t # Elapsed time is how many delta_t's elapsed * delta_t so steps * delta_t

half_times = []

for _ in range(num_simulations): # Run the simulation for sum_simulations times
    half_time = homogenous(N0, lambda_0, delta_t)
    half_times.append(half_time)

half_times = np.array(half_times)

mean_half_time = np.mean(half_times) # Compute the mean of the half times across all simulations
variance_half_time = np.var(half_times) # Compute the variance of the half times across all simulations

print(f"Mean half time: {mean_half_time} seconds")
print(f"Variance of half time: {variance_half_time} seconds^2")
```

Part a) answer:

Mean of the half times: $\textbf{1.6722}$ seconds.

Variance of the half times $\textbf{0.00057793}$ seconds^2.

$\textbf{Part b)}$

```python
def inhomogenous(N0, lambda_0, delta_t):
    N_t = N0
    steps = 0
    while N_t > N0 / 2:
        m = np.random.poisson(lambda_0 * N_t * delta_t)
        N_t -= m # decrese the number of nucleii by the number of decays at this step
        steps += 1 # increment the number of steps taken so far (one delta_t elapsed in time)
    return steps * delta_t # Elapsed time = steps * delta_t

half_times = []

for _ in range(num_simulations): # Run num_simulations
    half_time = inhomogenous(N0, lambda_0, delta_t)
    half_times.append(half_time)

half_times = np.array(half_times)

mean_half_time = np.mean(half_times) # Compute the mean of the half times across all simulations
variance_half_time = np.var(half_times) # Compute the variance of the half times across all simulations

print(f"Mean half time: {mean_half_time} seconds")
print(f"Variance of half time: {variance_half_time} seconds^2")
```

Part b) answer:

Mean of the half times: $\textbf{2.3104}$ seconds.

Variance of the half times: $\textbf{0.0011012}$ seconds^2.
